# # Data Cleaning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv("C:\\Users\\SAYAN DAS\\OneDrive\\Desktop\\SkillCraft\\T2\\Titanic-Dataset.csv")

# missing 'Age' values with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Droping columns with too many missing values or less relevance
df.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)

# Droping rows where 'Embarked' is missing
df.dropna(subset=['Embarked'], inplace=True)

# ----------------------------------------
# Step 2: Converting categorical columns to numeric
# ----------------------------------------

# One-hot encode categorical features
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

```

```python
#Checking
df.info()
```

## Visual Patterns and Insights

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Survival count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()


```

```python
# Survival by Sex
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Sex')
plt.show()

```

```python
# Age distribution by Survival
sns.histplot(data=df, x='Age', hue='Survived', kde=True, bins=30)
plt.title('Age Distribution by Survival')
plt.show()

```

```python
# Class vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Passenger Class vs Survival')
plt.show()
```
